```python
import numpy as np
import pandas as pd
```

```python
def make_data(species,start=np.datetime64('2024-01-01T00:00:00') ,end=np.datetime64('2025-12-31T23:59:59'),n=100, seed=None):    
    if seed is not None:
        np.random.seed(seed)

    df = pd.DataFrame({
        "date": pd.to_datetime(start + (end - start) * np.random.rand(n)),
        "camId": np.random.randint(1, 4, size=n),
        "species": np.random.choice(species, size=n),
        "count": np.random.randint(1, 11, size=n),
        "confidence": np.round(np.random.uniform(50, 100, size=n), 2)
    })

    return df

species = ['Tiger','Lion','Bear','Elephant','Dear']
df = make_data(species=species,n=2000,seed=42)
df.head()
```

```python
end = np.datetime64('2025-12-31T23:59:59')
start = end - np.timedelta64(7, "D")
print(start)
df1 = make_data(species=species,n=40,seed=42,start=start,end=end)
df2 = make_data(species=species,n=20,seed=42,start=end - np.timedelta64(1,"D"), end=end)
df2.head()
```

```python
df = pd.concat([df, df1, df2], ignore_index=True)
```
